## Imports

```python
## TODO
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

## Load data

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>

```python
## TODO
df = pd.read_excel('worldcities.xlsx')
df
```

 ## Add column

<div dir="rtl">

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
در این مرحله با توجه به ستون جمعیت یک ستون جدید تحت عنوان Population Level   اضافه کنید به طوری که به 4 دسته زیر  بر اساس وضعیت هر داده اختصاص داده می شودند.
</br>
    <div dir="ltr">
Low: 0-25%
    </br>
Mid: 25-50%
    </br>
High: 50-70%
    </br>
Over: 70-100%
    </div>
</font>
</p>
</div>

```python
## TODO
df['population_level'] = pd.qcut(df['population'], q=[0, 0.25, 0.5, 0.70, 1.0], labels=['Low', 'Mid', 'High', 'Over'])
df
```

## Preproccess

<div dir="rtl">

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    در این بخش باید ستون های خالی داده ها راحذف کرده و در نهایت تمامی ستون ها را به داده های عددی تبدیل کنید به جز ستون اضافه شده
</font>
</p>
</div>

```python
# TODO
df = df.dropna()
df_encoded = df.copy()

# Get the non-numeric column names
non_numeric_cols = df_encoded.select_dtypes(exclude=['number']).columns.tolist()
non_numeric_cols.remove('population_level')  # Exclude 'population_level'

# Apply label encoding to non-numeric columns
label_encoder = LabelEncoder()
for col in non_numeric_cols:
    df_encoded[col] = label_encoder.fit_transform(df_encoded[col])
df_encoded
```

<div dir="rtl">

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
 حال باید داده های به دست آمده از قسمت قبلی را نرمال سازی کنید.
</font>
</p>
</div>

```python
# TODO
df_standardized = df_encoded.copy()


numerical_cols = df_standardized.select_dtypes(include=['number']).columns.tolist()


scaler = StandardScaler()
df_standardized[numerical_cols] = scaler.fit_transform(df_standardized[numerical_cols])
df_standardized
```

## Split

<div dir="rtl">

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
داده ها را به دو بخش تست و آموزش با نسبت 80 به 20 تقسیم کنید.
</font>
</p>
</div>

```python
# TODO
train_df, test_df = train_test_split(df_standardized, test_size=0.2, random_state=42)
```

## Regression

<div dir="rtl">

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
برای این قسمت باید ستون Population  را به عنوان برچسب درنظر گرفته و از بقیه ستون هابه عنوان فیچر استفاده کنید
</font
</p>
</div>

<div dir="rtl">
<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
باید از رگرسیون خطی
،درجه 1 و درجه 2 استفاده کرده و در هر مورد دقت و خطای میانگین مربعات را بدست آورده و نتایج را باهم مقایسه کنید.</font
</p>
    
</div>

```python
train_df
```

```python
# TODO
X_train = train_df.drop(columns=['population', 'population_level'])
X_test = test_df.drop(columns=['population', 'population_level'])
y_train = train_df['population'] 
y_test = test_df['population']


linear_reg = LinearRegression()
linear_reg.fit(X_train, y_train)

y_pred = linear_reg.predict(X_test)
mse_first_order = mean_squared_error(y_test, y_pred)


from sklearn.preprocessing import PolynomialFeatures


poly = PolynomialFeatures(degree=2)
X_poly_train = poly.fit_transform(X_train)
X_poly_test = poly.fit_transform(X_test)


linear_reg_2nd_order = LinearRegression()
linear_reg_2nd_order.fit(X_poly_train, y_train)


y_pred_2nd_order = linear_reg_2nd_order.predict(X_poly_test)

mse_2nd_order = mean_squared_error(y_test, y_pred_2nd_order)
print(f'the first order loss is :{mse_first_order}')
print(f'the second orser loss is: {mse_2nd_order}')
```

### Decision Tree (Entropy)

```python
train_df
```

```python
# TODO
y_train = train_df['population_level']
y_test = test_df['population_level']

dt_entropy_model = DecisionTreeClassifier(criterion='entropy', random_state=42)
dt_entropy_model.fit(X_train, y_train)

y_pred_entropy = dt_entropy_model.predict(X_test)
accuracy_entropy = accuracy_score(y_test, y_pred_entropy)
conf_matrix_entropy = confusion_matrix(y_test, y_pred_entropy)
classification_report_entropy = classification_report(y_test, y_pred_entropy)


print("Decision Tree using Entropy:")
print(f"Accuracy: {accuracy_entropy}")
print("Confusion Matrix:")
print(conf_matrix_entropy)
print("Classification Report:")
print(classification_report_entropy)
```

### Random Forest (Entropy)

```python
# TODO
rf_entropy_model = RandomForestClassifier(criterion='entropy', random_state=42)
rf_entropy_model.fit(X_train, y_train)


y_pred_rf_entropy = rf_entropy_model.predict(X_test)


accuracy_rf_entropy = accuracy_score(y_test, y_pred_rf_entropy)
conf_matrix_rf_entropy = confusion_matrix(y_test, y_pred_rf_entropy)
classification_report_rf_entropy = classification_report(y_test, y_pred_rf_entropy)


print("Random Forest using Entropy:")
print(f"Accuracy: {accuracy_rf_entropy}")
print("Confusion Matrix:")
print(conf_matrix_rf_entropy)
print("Classification Report:")
print(classification_report_rf_entropy)
```

### KNN K=2

```python
# TODO
knn_model = KNeighborsClassifier(n_neighbors=2)
knn_model.fit(X_train, y_train)


y_pred_knn = knn_model.predict(X_test)


accuracy_knn = accuracy_score(y_test, y_pred_knn)
conf_matrix_knn = confusion_matrix(y_test, y_pred_knn)
classification_report_knn = classification_report(y_test, y_pred_knn)


print("K-Nearest Neighbors (KNN) with K=2:")
print(f"Accuracy: {accuracy_knn}")
print("Confusion Matrix:")
print(conf_matrix_knn)
print("Classification Report:")
print(classification_report_knn)
```
